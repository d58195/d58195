```python
data = np.array([[5,15,25,35,45,55],[5,20,14,32,22,38]])
xdata = data[0]
ydata = data[1]

N=len(xdata)
X=sum(xdata)/N
Y=sum(ydata)/N
X2=sum([i**2 for i in xdata])/N
XY=sum(xdata*ydata)/N

a2,b2=(XY-X*Y)/(X2-X*X),(X2*Y-X*XY)/(X2-X*X)
a2,b2
```

```python
plt.plot(data[0],data[1],'o',label='exp');
plt.xlabel('x');plt.ylabel('y')

x=np.arange(0,60,0.1)
y=a2*x+b2
plt.plot(x,y,'k-',label='fitted line');
plt.legend();
```

```python
x = 5
x_old = 2 #initial
eps = 0.01
precision = 0.000001

# 도함수 (미분)
def fp(x):
    return 2*x-1

# 경사하강법
while (abs(x - x_old)>precision):
    x_old = x
    x = x_old - eps*fp(x_old)
```
